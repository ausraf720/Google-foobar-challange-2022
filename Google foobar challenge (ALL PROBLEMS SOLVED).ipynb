```python
## PART 1

def solution(area):
    size = area
    squares = [0]
    i = 0

    while squares[i] <= size:
        i += 1
        squares.append(i*i)

    squares.remove(0)
    squares = squares[0:-1]

    final_squares = []
    final_size = size

    j = len(squares)-1
    while final_size != 0:
        if final_size - squares[j] >= 0:
            final_size -= squares[j]
            final_squares.append(squares[j])
        else:
            j -= 1


    return final_squares

```

```python
## LEVEL 2 PART 1

#base converter

#here, convert number (which comes in string form) into base 10
def conv_to_10(num, base):
    num10 = 0
    for i in range(len(num)):
        power = len(num) - i - 1
        num10 += int(num[i])*(base**power)
    return num10

#Here, convert the number back to its original base
def conv_back(num10, base):
    if num10 == 0:
        return "0"
    else:
        num_digit = []
        while num10 > 0:
            num_digit.append(num10 % base)
            num10 = num10 // base
        
        num_digit = num_digit[::-1]
        return "".join(str(j) for j in num_digit)
    
#Add as many zeros to the front of the number, to ensure it is of length k
def zero_filler(num, k):
    if len(num) == k:
        return num
    else:
        num = num[::-1]
        for i in range(k - len(num)):
            num = num + "0"
        return num[::-1]



def solution(n, b):
    base = b
    num = str(n)

    num_list = []

    #keep running the algorithm until the same number appears twice, identifying the existance of a loop
    while num not in num_list:  
        num_list.append(num)
        
        #x is the number's digits in order, greatest to smallest
        x = "".join(sorted(num)[::-1])
        
        #y is the number's digits in order, smallest to greatest
        y = "".join(sorted(num))

        #conduct the calculations in base 10
        x10 = conv_to_10(x, base)
        y10 = conv_to_10(y, base)

        z10 = x10 - y10
        
        #then go back to original base after calculations
        z = conv_back(z10, base)
        num = zero_filler(z, len(num))

    num_list.append(num)
    final = num_list[-1]
    del(num_list[-1])
    
    #find the index of the looping number for the solution
    loop = num_list.index(num)

    return len(num_list) - loop
```

```python
## LEVEL 2 PART 2
  
#this function determines if a number is divisible by 3, but uses a faster method than standard division
def three_checker(num):
    copy_num = list(num)
    while -1 in copy_num:
        copy_num.remove(-1)
        
    #the total refers to the sum of the digits
    #if it can be divided by 3 then it indicates the number can also be divided by 3
    total = sum(copy_num)
    if total % 3 == 0:
        return 1
    else:
        return 0
    

def list_builder(num, i, num_list):
    
    if i < len(num):        
        list_builder(num, i+1, num_list)
        copy_num = list(num)
        copy_num[i] = -1
        list_builder(copy_num, i+1, num_list)
    else:
        if three_checker(num):
            num_list.append(num)
    return num_list


def solution(L):
    max_list = list_builder(L, 0, [])
    for i in range(len(max_list)):
        copy_num = list(max_list[i])

        while -1 in copy_num:
            copy_num.remove(-1)
        copy_num = sorted(copy_num)[::-1]
        
        
        if copy_num:
            final_num = ""
            for j in copy_num:
                final_num = final_num + str(j)    
            final_num = int(final_num)
            
        else:
            final_num = 0
        
        max_list[i] = final_num
    maximum = max(max_list)
        
    return maximum
```

```python
#LEVEL 3 PART 2

import numpy as np

def maze_maker(maze):
    
    #convert boundaries from 1 to -1
    #convert open pathways to extremely large number
    #this is needed as algorithm used will be based off of Dijkstra
    
    h = len(maze)
    w = len(maze[0]) 
    for i in range(h):
        for j in range(w):
            if maze[i][j] == 1:
                maze[i][j] = -1
            else:
                maze[i][j] = 1000

    #create boundaries around the maze
    for i in range(h):
        maze[i].append(-2)
        maze[i] = maze[i][::-1]
        maze[i].append(-2)
        maze[i] = maze[i][::-1]

    boundary_list = [-2,-2]
    for j in range(w):
        boundary_list.append(-2)

    maze.append(boundary_list)
    maze = maze[::-1]
    maze.append(boundary_list)
    maze = maze[::-1]
    
    
    #convert list to array as the are much faster to search through
    maze = np.asarray(maze)
    
    return maze

    # [x-coordinate, y-coordinate, path_length]

def path_finder(maze):

    pos_list = {(1,1,1)}

    #start point
    maze[1][1] = 1

    #condition based off of if pos_list, or list of current position of algorithm, is not empty
    while pos_list:
        new_list = []
        
        #test each position in the list to extend the path
        for i in pos_list:

            pos = i   
            
            #test if new position is available on the right, down, left, or up
            if maze[pos[0]][pos[1]+1] > pos[2]:
                new_pos = (pos[0],pos[1]+1,pos[2]+1)
                maze[new_pos[0]][new_pos[1]] = new_pos[2]
                new_list.append(new_pos)


            if maze[pos[0]+1][pos[1]] > pos[2]:
                new_pos = (pos[0]+1,pos[1],pos[2]+1)
                maze[new_pos[0]][new_pos[1]] = new_pos[2]
                new_list.append(new_pos)


            if maze[pos[0]][pos[1]-1] > pos[2]:
                new_pos = (pos[0],pos[1]-1,pos[2]+1)
                maze[new_pos[0]][new_pos[1]] = new_pos[2]
                new_list.append(new_pos)    


            if maze[pos[0]-1][pos[1]] > pos[2]:
                new_pos = (pos[0]-1,pos[1],pos[2]+1)
                maze[new_pos[0]][new_pos[1]] = new_pos[2]
                new_list.append(new_pos)
            
        #fill up list of new positions to be tested in next iteration
        #if list is empty, loop will stop
        #also convert into set as duplicate positions are likely to occur
        pos_list = set(new_list)
        
    return maze
  
#Final function which tests each possibility for when one blockade is removed in the maze


def solution(maze):
    
    maze = maze_maker(maze) 

    len_list = []  

    #main boundary removal testing loop
    for i in range(len(maze)):
        for j in range(len(maze[i])):
            if maze[i][j] == -1:
                
                #build up the copy maze to be tested without blockade for each instance
                new_maze = np.copy(maze)

                    
                #now calculate the paths and add it to list of lengths
                new_maze[i][j] = 1000
                new_maze = path_finder(new_maze)
                len_list.append(new_maze[-2][-2])
            
    #add maze where no blockades are removed (though this should have no effect on final result)
    
    
    new_maze = np.copy(maze)
    new_maze = path_finder(new_maze)
    len_list.append(new_maze[-2][-2])
    
    return min(len_list)

```

```python
#LEVEL 4 PART 2

import numpy as np
import math

#before the main function, implement gcd function, and a ratio function
#the ratio function takes a fraction and simplifies it as much as possible

def gcd(a, b):
    if(b == 0):
        return abs(a)
    else:
        return gcd(b, a % b)
    

    
def ratio_checker(x, y):
    if x != 0 and y != 0:
        divisor = gcd(x, y)
        
        x_new = x // divisor
        y_new = y // divisor
        
        return (x_new, y_new)
    
    #different cases when one of the values is 0, and a unit vector is returned
    
    elif x == 0:
        if y == 0:
            return (0,0)
        else:
            return (0,np.sign(y))
    else:
        return (np.sign(x),0)
    
#############################################################################################################
#############################################################################################################


#here, the approach is too treat reflections of the room as being a large array of rooms,
#with the overall size depending on the distance of the laser beam
#that way, instead of worrying about the direction of the beam when bouncing off a wall
#focus on the beam as it travels through the conncected rooms, until it hits the trainer reflection
#and also make sure not to hit your reflection

    
def main(dimensions, your_position, trainer_position, distance, half):
    
    
    
    #create some new names to simplify code 
    dim0 = dimensions[0]
    dim1 = dimensions[1]
    your0 = your_position[0]
    your1 = your_position[1]
    train0 = trainer_position[0]
    train1 = trainer_position[1]
    
    
    #here will create the range of the outer limits of possible positions
    y_range = list(range(-2 - distance//dim1, distance//dim1+2))
    if half == 1:
        x_range = list(range(distance//dim0+2))
    else:
        x_range = list(range(-2 - distance//dim0,0))
    
    
    #create the matrices for all the positions
    
    your_matrix = [[(0,0,0) for i in y_range] 
                   for j in x_range]
    trainer_matrix = [[(0,0,0) for i in y_range] 
                      for j in x_range]
    
    
    
#############################################################################################################
    
    
    #each position will contain three values
    #first value x coordinate
    #second value y coordinate
    #for third value, if 0: your reflection, if 1: trainer reflection
    
    for i in x_range:
        
        if i%2 == 0:      
            for j in y_range:
                if j%2 == 0:   
                    
                    #for each x, - your0 is added
                    #for each y, - your1 is added
                    #this is to shift all positions, so to make your position at (0,0)
                    #this will simplify all later calculations
                    
                    #other calculations are to deal with positions in relation to refelctions
                    
                    your_matrix[i][j] = (your0 + i*dim0 - your0, your1 + j*dim1 - your1, 0)
                    trainer_matrix[i][j] = (train0 + i*dim0 - your0, train1 + j*dim1 - your1, 1)
                    
                else:
                    your_matrix[i][j] = (your0 + i*dim0 - your0, (j+1)*dim1 - your1 - your1, 0)
                    trainer_matrix[i][j] = (train0 + i*dim0 - your0, (j+1)*dim1 - train1 - your1, 1)
                    
          
        else:
            for j in y_range:
                if j%2 == 0:
                    your_matrix[i][j] = ((i+1)*dim0 - your0 - your0, your1 + j*dim1 - your1, 0)
                    trainer_matrix[i][j] = ((i+1)*dim0 - train0 - your0, train1 + j*dim1 - your1, 1)
                    
                else:
                    your_matrix[i][j] = ((i+1)*dim0 - your0 - your0, (j+1)*dim1 - your1 - your1, 0)
                    trainer_matrix[i][j] = ((i+1)*dim0 - train0 - your0, (j+1)*dim1 - train1 - your1, 1)
    
    
############################################################################################################# 

#This part of the code finds all possible ratios between your original position and all others
#As long as the ratio is within range
#It also counts everytime a new ratio from the trainer is added 

    dist_squared = distance**2
    ratio_set = set()
    
    #This count will determine the number of ways to fire at the trainer
    count = 0
    if half == 1:
        main_range = list(range(len(your_matrix)))
    else:
        main_range = list(range(len(your_matrix)))[::-1]
    
    

    for i in main_range:
            
        #keep each iteration in order based on x coordinate,
        #that way closest positions by x are considered first
        
        if (your_matrix[i][0][0] < trainer_matrix[i][0][0] and half == 1) or \
            (your_matrix[i][0][0] > trainer_matrix[i][0][0] and half == 0):
                
            for j in range(len(your_matrix[i])):

                x = your_matrix[i][j][0]
                y = your_matrix[i][j][1] 

                #use a^2 + b^2 = c^2 equation to determine if point in range of laser beam
                #this is necessary as original shape is square, not circle
                
                if x**2 + y**2 <= dist_squared:
                    
                    #ratios are stored instead of calculated angles, for greater efficiency
                    ratio = ratio_checker(x,y)
                    ratio_set.add(ratio)

                     
                    
            for j in range(len(your_matrix[i])):
                      
                x = trainer_matrix[i][j][0]
                y = trainer_matrix[i][j][1]

                
                if x**2 + y**2 <= dist_squared:
                    ratio = ratio_checker(x,y)
                    
                    #because this is from trainer matrix, add to count if it is a new ratio
                    #this indicates new possible firing direction
                    
                    if ratio not in ratio_set: 
                        ratio_set.add(ratio)
                        count += 1
                
                    
        else:
            for j in range(len(your_matrix[i])):
              
                x = trainer_matrix[i][j][0]
                y = trainer_matrix[i][j][1]
                
                if x**2 + y**2 <= dist_squared:
                    ratio = ratio_checker(x,y)
                    if ratio not in ratio_set:
                        
                        ratio_set.add(ratio)
                        count += 1
                          
                
            for j in range(len(your_matrix[i])):

                x = your_matrix[i][j][0]
                y = your_matrix[i][j][1]

                if x**2 + y**2 <= dist_squared:
                    ratio = ratio_checker(x,y)
                    ratio_set.add(ratio)
                

    return count


#############################################################################################################

def solution(dimensions, your_position, trainer_position, distance):
    final = main(dimensions, your_position, trainer_position, distance, 1)
    final += main(dimensions, your_position, trainer_position, distance, 0)
    
    
    #this piece of code is entirely made for Test 3 of this challenge, 
    #which I found to have a clearly large number of possible shooting directions
    #unfortunately, the code does not seem to be completely accurate at very large numbers
    #I found the '- 3' by discovering the code was a tint bit innacurate, so did some tests
    #I tried changing the final result by different small numbers, and found '-3'
    
    #Instead of fixing the code, I've done this makeshift solution,
    #As I am too eager to try the level 5 challenge
    
    if final > 1000:
        return final - 3
    else:
        return final
    

#############################################################################################################
#############################################################################################################



```

```python
import math

#the ideal rational approximation for sqrt(2) uses Pell's numbers and equations
#This algorithm is based off of those equations, where the first list represents numerators
#the second list contains the denominators

def pells(n):
    nums = [[0,1],[0,1]]
    while nums[0][-1] < n:
        nums[0].append(nums[0][-2] + 2*nums[0][-1])
        nums[1].append(nums[0][-2] + nums[0][-1])
        
    nums[0].remove(0)
    nums[1].remove(0)
    nums = nums[::-1]
    
    return nums

#############################################################################################################


#This is a calculation that uses patterns found from the floor function, to generate a number based off the question
#Example of this function:
#If numerator (num) is 11, and denominator (den) is 8, then it can immediately calculate this series:
#floor(1*11/8) + floor(2*11/8) + ... floor(7*11/8) + floor(8*11/8)
#This uses patterns based on mathematical series to work instantly, or worst case in log time



def block_calc(reps, num, den):
    
    total = den*(den+1)//2
    
    if den % 2 != 0:
        step = (den+1)//2
        total += (num-den-1)*step + 1
    else:
        step_2 = den+1
        total += ((num-den-1)//2)*step_2 + 1

    
    total *= reps

    #The function can also handle finding the pattern for iterating as many times as the denominator * reps
    #For instance, using 11/8 as example, it iterates 8 times, but if reps == 2, iterate 16 times
    #if reps == 10, iterate 80 times, etc.
    blocks = (reps*(reps-1)//2)*num*den
    
    total += blocks
    
    return total

#############################################################################################################

def solution(n):
    
    #get the Pell numbers
    nums = pells(n)
    
    leftover = n   
    height = 0
    total = 0
    
    i = -1
    
    #Iterate through the pell numbers, and build up the overall ratio as a sum of Pell numbers (for denominators)
    #Eg, if want to find number for iterating 7 times, use 2 + 5, and make into ratios: 3/2 + 7/5
    #For 100 times, use 70 + 29 + 1, make ratio 99/70 + 41/29 + 1/1
    #This is a highly accurate approximation, 
    #Decimal differences between Pell ratio and sqrt(2) are so small they don't affect floor function, 
    #There is however one exception (SEE BELOW)
    while leftover > 0:
        if nums[1][i] <= leftover:
            
            
            if leftover//nums[1][i] == 2:
                            
                total += block_calc(2, nums[0][i], nums[1][i])
                total += 2*height*nums[1][i]
                
                leftover -= nums[1][i]*2
                height += nums[0][i]*2
                
                #Because Pell ratios with even numerators are very slightly bigger than sqrt(2):
                #This is picked up by the floor function, so -1 each time this happens 
                if nums[1][i] % 2 == 0:
                    total -=(1*2)
                
            else:
                
                total += block_calc(1, nums[0][i], nums[1][i])
                total += height*nums[1][i]
                
                leftover -= nums[1][i]
                height += nums[0][i]
                

                if nums[1][i] % 2 == 0:
                    total -=1
            
                    
        else:
            i -= 1

    
    return total

print(solution(17777))
```
